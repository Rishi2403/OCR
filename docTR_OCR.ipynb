```python
def extract_values(d):
    if isinstance(d, dict):
        for key, value in d.items():
            if key == 'value' and value is not None and value is not 0:
                print(value)  # Print the 'value' field if it's not None
            else:
                extract_values(value)
    elif isinstance(d, list):
        for item in d:
            extract_values(item)

# Extract values
extract_values(json_respose)
```

```python
import re
import json

# Assuming json_respose is a dictionary or JSON object, convert it to a string
json_string = json.dumps(json_respose)

# Define your regex pattern
watt_pat = r"(\d+(\.\d+)?\s*(kilovolt|volt|millivolt|kv|v|mv|mg|cm))"

# Now, apply re.findall() on the string representation of the JSON
watts = re.findall(watt_pat, json_string)

# Check if any watts were found
if watts:
  for data in watts:
    print(data[0])
    #print(watts)
else:
    print("No wattage found")
```

```python
import re

# Your entity_unit_map dictionary
entity_unit_map = {
    'width': {'centimetre', 'foot', 'inch', 'metre', 'millimetre', 'yard'},
    'depth': {'centimetre', 'foot', 'inch', 'metre', 'millimetre', 'yard'},
    'height': {'centimetre', 'foot', 'inch', 'metre', 'millimetre', 'yard'},
    'item_weight': {'gram', 'kilogram', 'microgram', 'milligram', 'ounce', 'pound', 'ton'},
    'maximum_weight_recommendation': {'gram', 'kilogram', 'microgram', 'milligram', 'ounce', 'pound', 'ton'},
    'voltage': {'kilovolt', 'millivolt', 'volt'},
    'wattage': {'kilowatt', 'watt'},
    'item_volume': {'centilitre', 'cubic foot', 'cubic inch', 'cup', 'decilitre', 'fluid ounce', 'gallon', 'imperial gallon', 'litre', 'microlitre', 'millilitre', 'pint', 'quart'}
}

# Convert the dictionary into a string for regex extraction
data_string = str(entity_unit_map)

# Define the regex pattern to match unit names
pattern = r'\b[a-zA-Z ]+\b'

# Find all matches in the string
matches = re.findall(pattern, data_string)

# Remove duplicates by converting to a set
unique_units = set(matches)

# Print the extracted unit values
print(unique_units)
```
